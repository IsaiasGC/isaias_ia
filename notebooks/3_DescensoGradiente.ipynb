```python
import numpy as np
import scipy as sc

import matplotlib.pyplot as plt
```

```python
func = lambda th: np.sin(1 / 2 * th[0] ** 2 - 1/4 * th[1] ** 2 + 3) * np.cos(2 * th[0] + 1 - np.e ** th[1])
```

```python
res = 1000

_X = np.linspace(-3, 3, res)
_Y = np.linspace(-3, 3, res)

_Z = np.zeros((res, res))

for ix, x in enumerate(_X):
  for iy, y in enumerate(_Y):
    _Z[ix, iy] = func([x,y])
```

```python
Theta = np.random.rand(2) * 4 -2

h = 0.005
lr = 0.05

grad = np.zeros(2)
```

```python
plt.contourf(_X, _Y, _Z, 500)
plt.colorbar()
print("Theta: ",Theta)
plt.plot(Theta[0], Theta[1], "o", c="red")

for _ in range(1000):
  for it, th in enumerate(Theta):
    _T = np.copy(Theta)
    _T[it] = _T[it] + h
    deriv = (func(_T) - func(Theta)) / h

    grad[it] = deriv

  Theta = Theta - lr * grad

plt.plot(Theta[0], Theta[1], "o", c="green")
plt.show()
```
